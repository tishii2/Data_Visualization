```python
import pandas as pd
import numpy as np
```

```python
#Deleted the first 4 rows in csv file so these became column names
df = pd.read_csv('2-4-2.csv')
df.columns
df.fillna(0)
```

```python
print(df['Murder'])
```

```python
print(df['Robbery'])
```

```python
df['Robbery'] = df['Robbery'].replace('0', 0, inplace = True)
df['Robbery'].replace(df['Robbery'] != '0', 1, inplace=True)
```

df['Robbery'] = (df['Robbery'] > 0).astype(int)
df['Murder'] = (df['Murder'] > 0).astype(int)

```python
df['Murder'] = df['Murder'].replace('0.0', 0, inplace = True)
df['Murder'].replace(df['Murder'] != '0.0', 1, inplace=True)
```
