```python
print ('\033[1m' + 'THE NUMBERS')
print('\033[0m')


print('\033[1m' + 'Michael Jordan' + '\033[0m' + ' Career Stats')
mj = pd.read_csv("jordan_career.csv")
print(mj)
```

```python
print('\033[1m' + 'Lebron James' + '\033[0m' + ' Career Stats')
bron = pd.read_csv("lebron_career.csv")
print(bron)
```

```python
print('Here you can see and compare raw data between Michael Jordan and Lebron James. This is best for when you want to look at statistics from a specific game. Not so great in comparing the two.')
```

```python
print ('\033[1m' + '1.) Who is the better player based on the main basketball statistics? (Points, Rebounds, Assists)')
print('\033[0m')

plt.figure(figsize=(25, 7))
plt.subplot(1, 2, 1)
plt.scatter(
    x=mj['game'], 
    y=mj['pts'],
    color='black',
    marker='x', s=5,
)
plt.title('Michael Jordan Career Points')
plt.xlabel("Game # (1 - 1072)")
plt.ylabel("Points")
plt.ylim([0, 70])

plt.subplot(1, 2, 2)
plt.scatter(
    x=bron['game'], 
    y=bron['pts'],
    color='blue',
    marker='x', s=5,
)
plt.title("Lebron James Career Points")
plt.xlabel("Game # (1 - 1143)")
plt.ylabel("Points")
plt.ylim([0, 70])
plt.show()

print('Here you can distinguish how both of the players scored throughout their career (so far). With Michael Jordan playing a total of 1072 games, and Lebron James playing a total of 1143 games so far, their scoring numbers tend to match up pretty evenly amongst one another. Although both impressive, Lebron\'s visual looks much more consistant and show signs of improvement from here on out as opposed to Michael\'s in which his scoring marks dropped off during this point in his career.')
```

```python
plt.figure(figsize=(10, 5))
jordan = plt.hist(mj['trb'], color='black', bins=np.arange(0, 20), alpha=0.5)
james = plt.hist(bron['trb'], color='blue', bins=np.arange(0, 20), alpha=0.5)
plt.title("Rebounds Per Game Comparison")
plt.xlabel("Rebounds")
plt.ylabel("# of Games")
black_patch = mpatches.Patch(color='black', alpha=.5, label='Michael Jordan')
blue_patch = mpatches.Patch(color='blue', alpha=.5, label='Lebron James')
plt.legend(handles=[black_patch, blue_patch])
plt.show()
print('\nThis visual compares the number of games with a specific amount of rebounds both players had. Being overlapped on top of one another, it\'s easily distinguishable that Lebron James clearly has an advantage in rebound marks throughout his career. It\'s worth mentioning that Michael Jordan, a Shooting Guard is not as relied on as Lebron James\' position as Small Forward to gather rebounds on a day to day basis. But in this instance, we\'re comparing the players in terms of making their team\'s better in which Lebron clearly has the edge.')
```

```python
plt.figure(figsize=(10, 5))
jordan = plt.hist(mj['ast'], color='black', bins=np.arange(0, 20), alpha=0.5)
james = plt.hist(bron['ast'], color='blue', bins=np.arange(0, 20), alpha=0.5)
plt.title("Assists Per Game Comparison")
plt.xlabel("Assists")
plt.ylabel("# of Games")
black_patch = mpatches.Patch(color='black', alpha=.5, label='Michael Jordan')
blue_patch = mpatches.Patch(color='blue', alpha=.5, label='Lebron James')
plt.legend(handles=[black_patch, blue_patch])
plt.show()

print('\nThis visual is a representation of both players\' assist numbers throughout their career. As a leader of any team, a major part of your responsibility is to make your teammates better and make their lives easier. In this instance, the ability to create easy, open shots for your teammates. Even in this comparison, you can tell that Lebron James\' ability to assist his teammates surpasses Michael Jordan\'s. Unlike the rebounds, assists are not a stat reliant on any specific position in which case this camparison proves Lebron\'s vision and ability to find the open man surpasses that of Jordan\'s.')
```
