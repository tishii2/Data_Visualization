```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bernoulli= np.random.binomial(1, .5, 100)
plt.hist(bernoulli)
plt.axvline(bernoulli.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(bernoulli.std(), color='grey', linestyle='solid', linewidth=2)
print('mean: ', bernoulli.mean())
print('standard deviation: ', bernoulli.std())
plt.show()
```

```python
binomial= np.random.binomial(20, .5, 100)
plt.hist(binomial)
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomial.std(), color='grey', linestyle='solid', linewidth=2)
print('mean: ', binomial.mean())
print('standard deviation: ', binomial.std())
plt.show()
```

```python
ecommerce = pd.DataFrame()
ecommerce['spending']= np.random.normal(75, 25, 100)

plt.hist(ecommerce['spending'])
plt.axvline(ecommerce['spending'].mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(ecommerce['spending'].std(), color='grey', linestyle='solid', linewidth=2)
print('mean: ', ecommerce.mean())
print('standard deviation: ', ecommerce.std())
plt.show()
```

```python
expo= np.random.exponential(10, 100)
plt.hist(expo)
plt.axvline(expo.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(expo.std(), color='grey', linestyle='solid', linewidth=2)
print('mean: ', expo.mean())
print('standard deviation: ', expo.std())
plt.show()
```

```python
normal1 = np.random.normal(5, .5, 100)
normal2 = np.random.normal(10, 1, 100)
normal3 = normal1 + normal2

mean = normal3.mean()
sd = np.std(normal3)

plt.hist(normal3)
plt.axvline(mean, color='b', linestyle='solid', linewidth=2)
plt.axvline(mean-sd, color='grey', linestyle='dashed', linewidth=2)
plt.axvline(mean+sd, color='grey',linestyle='dashed', linewidth=2)
plt.show()
```
