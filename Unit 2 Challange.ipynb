```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mj = pd.read_csv("jordan_career.csv")
print(mj)
```

```python
bron = pd.read_csv("lebron_career.csv")
print(bron)
```

```python
dame = pd.read_csv('dame_career.csv')
print(dame)
```

```python
plt.scatter(
    x=mj['game'], 
    y=mj['pts'],
    color='black',
    marker='x', s=5,
)
plt.show

#This visual represents Michael Jordan's career Points in a game marks. You can see in his 16 year career,
#he slowly declined consistency wise. 
```

```python
plt.scatter(
    x=bron['game'], 
    y=bron['pts'],
    color='blue',
    marker='x', s=5,
)
plt.show()

#This visual represents Lebron James' career Points in a game marks. During his past 15 years in the NBA,
#(Still playing) Lebron's Points in a game marks have been consistant and are on the rise again. 
```

```python
plt.scatter(
            x=dame['G'],
            y=dame['PTS'],
            color='red',
            marker='x', s=5,
)
plt.show()

#This visual is an extra I added of my hometown team's all-star, Damian Lillard. He has only been in the NBA 
#for6 years but his marks are impressive. 
```

```python
plt.figure(figsize=(25, 7))

plt.subplot(1, 3, 1)
plt.scatter(
    x=mj['game'], 
    y=mj['pts'],
    color='black',
    marker='x', s=5,
)
plt.title('Michael Jordan Career')
plt.xlabel("Game # (1 - 1072)")
plt.ylabel("Points")
plt.ylim([0, 70])

plt.subplot(1, 3, 2)
plt.scatter(
    x=bron['game'], 
    y=bron['pts'],
    color='blue',
    marker='x', s=5,
)
plt.title("Lebron James Career")
plt.xlabel("Game # (1 - 1143)")
plt.ylabel("Points")
plt.ylim([0, 70])

plt.subplot(1, 3, 3)
plt.scatter(
        x=dame['G'],
        y=dame['PTS'],
        color = 'red',
        marker='x', s=5
)

plt.title("Damian Lillard Career")
plt.xlabel("Game # (1-469)")
plt.ylabel("Points")
plt.xlim([-50, 1100])
plt.ylim([0, 70])

plt.show()

#This visual shows a side by side comparison of the three players and how their career points in a game mark 
#line up with one another. You can tell Lebron James' marks are far more consistant is still steadily improving. 
#Damian Lillard's marks are also impressive through his first 500 games and are similar to how Michael Jordan and 
#Lebron James started their careers. 
```
