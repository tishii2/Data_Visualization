# 1.) Who is the better player based on the main basketball statistics? (Points,  Assists, Field Goal Percentage)

## Points

```python
plt.figure(figsize=(25, 7))
plt.subplot(1, 2, 1)
plt.scatter(
    x=mj['game'], 
    y=mj['pts'],
    color='red',
    marker='x', s=5
)
plt.title('Michael Jordan Career Points')
plt.xlabel("Game # (1 - 1072)")
plt.ylabel("Points")
plt.ylim([0, 70])

plt.subplot(1, 2, 2)
plt.scatter(
    x=bron['game'], 
    y=bron['pts'],
    color='blue',
    marker='x', s=5,
)
plt.title("Lebron James Career Points")
plt.xlabel("Game # (1 - 1143)")
plt.ylabel("Points")
plt.ylim([0, 70])
plt.show()
```

Here you can distinguish how both of the players scored throughout their career (so far). With Michael Jordan playing a total of 1072 games, and Lebron James playing a total of 1143 games so far, their scoring numbers tend to match up pretty evenly amongst one another. Although both impressive, Lebron's visual looks much more consistant and show signs of improvement from here on out as opposed to Michael's in which his scoring marks dropped off during this point in his career.

## Assists

```python
plt.figure(figsize=(10, 5))
jordan = plt.hist(mj['ast'], color='red', bins=np.arange(0, 20), alpha=0.5)
james = plt.hist(bron['ast'], color='blue', bins=np.arange(0, 20), alpha=0.5)
plt.title("Assists Per Game Comparison")
plt.xlabel("Assists")
plt.ylabel("# of Games")
black_patch = mpatches.Patch(color='red', alpha=.5, label='Michael Jordan')
blue_patch = mpatches.Patch(color='blue', alpha=.5, label='Lebron James')
plt.legend(handles=[black_patch, blue_patch])

plt.axvline(mj['ast'].mean(), color='red', alpha=0.5, linestyle='solid', linewidth=2)
plt.axvline(bron['ast'].mean(), color='blue', alpha=0.5, linestyle='solid', linewidth=2)

print('Micheal Jordan Assist Average: ', + mj['ast'].mean())
print('Lebron James Assists Average: ', + bron['ast'].mean())
plt.show()
```

This visual is a representation of both players' assist numbers throughout their career. As a leader of any team, a major part of your responsibility is to make your teammates better and make their lives easier. In this instance, the ability to create easy, open shots for your teammates. Even in this comparison, you can tell that Lebron James' ability to assist his teammates surpasses Michael Jordan's. Unlike the rebounds, assists are not a stat reliant on any specific position in which case this camparison proves Lebron's vision and ability to find the open man surpasses that of Jordan's.

## Field Goal Percentage

```python
plt.figure(figsize=(10, 5))
jordan = plt.scatter(x=mj['game'],
                     y=mj['fgp'], 
                     color='red',
                     alpha=0.5
)
james = plt.scatter(x= bron['game'],
                    y=bron['fgp'],
                    color='blue',
                    alpha=0.5
)

print('Micheal Jordan Career Field Goal Percentage: ', + mj['fgp'].mean())
print('Lebron James Career Field Goal Percentage: ', + bron['fgp'].mean())
plt.title("Field Goal Percentage Per Game Comparison")
plt.xlabel("Game #")
plt.ylabel("FGP %")
black_patch = mpatches.Patch(color='red', alpha=.5, label='Michael Jordan')
blue_patch = mpatches.Patch(color='blue', alpha=.5, label='Lebron James')
plt.legend(handles=[black_patch, blue_patch])
plt.axhline(mj['fgp'].mean(), color='red', alpha=0.5, linestyle='solid', linewidth=2)
plt.axhline(bron['fgp'].mean(), color='blue', alpha=0.5, linestyle='solid', linewidth=2)
plt.show()
```

The numbers representing through this scatter plot convey the efficiency in which both players scored at. although very similar, Lebron James' field goal percentage surpasses that of Michael's by a tiny margin. You may also recognize the development between Lebron James' career thus far is much more positive than that of Jordan's. This is not only a testimate to his growth into a superstar, but a more efficient scorer than Michael Jordan. 

# 2.) Who is the better player based on the main defensive basketball statistics? (Rebounds, Blocks, Steals)

## Rebounds

```python
plt.figure(figsize=(10, 5))
jordan = plt.hist(mj['trb'], color='red', bins=np.arange(0, 20), alpha=0.5)
james = plt.hist(bron['trb'], color='blue', bins=np.arange(0, 20), alpha=0.5)
plt.title("Rebounds Per Game Comparison")
plt.xlabel("Rebounds")
plt.ylabel("# of Games")
black_patch = mpatches.Patch(color='red', alpha=.5, label='Michael Jordan')
blue_patch = mpatches.Patch(color='blue', alpha=.5, label='Lebron James')
plt.legend(handles=[black_patch, blue_patch])
plt.axvline(mj['trb'].mean(), color='red', alpha=0.5, linestyle='solid', linewidth=2)
plt.axvline(bron['trb'].mean(), color='blue', alpha=0.5, linestyle='solid', linewidth=2)
print('Micheal Jordan Rebounds Average: ', + mj['trb'].mean())
print('Lebron James Rebounds Average: ', + bron['trb'].mean())
plt.show()
```

This visual compares the number of games with a specific amount of rebounds both players had. Being overlapped on top of one another, it's easily distinguishable that Lebron James clearly has an advantage in rebound marks throughout his career. It's worth mentioning that Michael Jordan, a Shooting Guard is not as relied on as Lebron James' position as Small Forward to gather rebounds on a day to day basis. But in this instance, we're comparing the players in terms of making their team's better in which Lebron clearly has the edge.

Comparing the two players' blocks marks, Jordan has a slight edge over Lebron while also giving up 2 inches to Lebron's 6 feet 8 inch stature. Jordan's ability to time and position himself to average more blocks than Lebron is an extremely impressive.

# 3.) How much does each player hinder their team's play? (Turnovers)

```python
plt.subplot(1, 2, 1)
labels = ['Minutes Played', 'Minutes Out']
labels1 = ['Minutes Played: 41,011 Min.', 'Minutes Out: 10,445 Min.']
sizes1 = [41011, 10445]
colors = ['red', 'grey']
plt.pie(sizes1, colors=colors, startangle=90, labels=labels1)
plt.legend(labels, loc=9)
plt.axis('equal')
plt.title('Michael Jordan')

plt.subplot(1, 2, 2)
labels = ['Minutes Played', 'Minutes Out']
labels2 = ['Minutes Played: 44,298 Min.', 'Minutes Out: 10,566 Min.']
sizes2 = [44298, 10566]
colors = ['blue', 'grey']
plt.pie(sizes2, colors=colors, startangle=90, labels=labels2)
plt.legend(labels, loc=9)
plt.axis('equal')
plt.title('Lebron James')
fig = plt.gcf()
fig.set_size_inches(15,8)
plt.show()
```

This is a representation of both players' number of minutes played throughout their career. Michael Jordan, finished with 41,011 minutes out of 51,456 minutes possible. Lebron James, so far in his career has played 44,298 minutes out of a potential 54,864 minutes. This will help in putting turnover numbers into perspective.

## Turnovers

```python
plt.figure(figsize=(10, 5))
jordan = plt.hist(mj['tov'], color='red', bins=np.arange(0, 15), alpha=0.5)
james = plt.hist(bron['tov'], color='blue', bins=np.arange(0, 15), alpha=0.5)
plt.title("Turnover Per Game Comparison")
plt.xlabel("Turnovers")
plt.ylabel("# of Games")
black_patch = mpatches.Patch(color='red', alpha=.5, label='Michael Jordan')
blue_patch = mpatches.Patch(color='blue', alpha=.5, label='Lebron James')
plt.legend(handles=[black_patch, blue_patch])
plt.axvline(mj['tov'].mean(), color='red', alpha=0.5, linestyle='solid', linewidth=2)
plt.axvline(bron['tov'].mean(), color='blue', alpha=0.5, linestyle='solid', linewidth=2)

print('Micheal Jordan Turnovers Average: ', + mj['tov'].mean())
print('Lebron James Turnoveres Average: ', + bron['tov'].mean())
plt.show()
```

```python
from scipy.stats import ttest_ind
print(ttest_ind(mj['tov'], bron['tov'], equal_var=False))
```

After looking at the ratio of the number of minutes played between both players, it's reasonable to say that Michael Jordan was better in taking control of the ball and limiting the amount of turnovers he committed.
